```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Khai báo danh sách tên cột chuẩn phục vụ cho việc ép cột ở Vấn đề 1
column_names = ["Id", "Name", "Age", "Weight", 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']

# Tạo danh sách dải cột tạm thời (15 cột) để đọc an toàn dòng số 17 (Tránh ParserError)
temp_names = [f"col_{i}" for i in range(15)]


# =============================================================================
# #Vấn đề 1:  Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề
# “Missing header in the csv file”
# =============================================================================
df = pd.read_csv("patient_heart_rate.csv", names=temp_names, header=None, engine='python')
df = df.iloc[:, :10]
df.columns = column_names

print("--- Dữ liệu sau Vấn đề 1 ---")
print(df.head())
```

```python
# =============================================================================
# Vấn đề 2: Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao
# gồm “Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột
# =============================================================================
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
print("--- Dữ liệu sau Vấn đề 2 ---")
print(df)
```

```python
# =============================================================================
# Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu.
# Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn “kg”
# =============================================================================
weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])
    if "lbs" in x[-3:]:
        x = x[:-3]
        float_x = float(x)
        y = int(float_x / 2.2)
        y = str(y) + "kgs"
        weight[i] = y

print("--- Dữ liệu sau Vấn đề 3 ---")
print(df)
```

```python
# =============================================================================
# vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có
# thể đưa ra là xóa bỏ
# =============================================================================
df.dropna(how="all", inplace=True)
print("--- Dữ liệu sau Vấn đề 4 ---")
print(df)
```

```python
# =============================================================================
# Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname,
# age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải
# dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.
# =============================================================================
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print("--- Dữ liệu sau Vấn đề 5 ---")
print(df)
```

```python
# =============================================================================
# Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII.
# Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác
# hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)
# =============================================================================
df.Firstname.replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)
print("--- Dữ liệu sau Vấn đề 6 ---")
print(df)
```

```python
# =============================================================================
# VẤN ĐỀ 7 & VẤN ĐỀ 8: Xử lý khuyết thiếu và Chuyển bảng ngang sang dọc (Melt)
# =============================================================================
print("=== KẾT HỢP VẤN ĐỀ 7 & VẤN ĐỀ 8 ===")

# --- PHẦN 1: BƯỚC KHẮC PHỤC LỖI CHẠY LẠI ---
# Nếu bấm nút chạy lại ô này, ta cần xóa các cột kết quả cũ để reset cấu trúc bảng
if 'PulseRate' in df.columns:
    df = df.drop(columns=['PulseRate'])
if 'sex_and_time' in df.columns:
    df = df.drop(columns=['sex_and_time'])


# --- PHẦN 2: XỬ LÝ VẤN ĐỀ 7 (Ép kiểu & Điền khuyết an toàn) ---
# Trích xuất chỉ lấy phần ký tự số, loại bỏ chữ kgs hoặc khoảng trắng thừa một cách triệt để
df['Weight'] = df['Weight'].astype(str).str.extract(r'(\d+)', expand=False)

# Ép kiểu số an toàn cho Tuổi và Cân nặng
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Tính toán giá trị trung bình để điền khuyết (Không xóa dòng bừa bãi làm trống bảng)
mean_age = df['Age'].mean()
if pd.isna(mean_age):
    mean_age = 30.0
df['Age'].fillna(mean_age, inplace=True)

mean_weight = df['Weight'].mean()
if pd.isna(mean_weight):
    mean_weight = 65.0
df['Weight'].fillna(mean_weight, inplace=True)

# Chỉ xóa dòng nếu cả Tuổi và Cân nặng đều trống hoàn toàn
df.dropna(subset=['Age', 'Weight'], how='all', inplace=True)

print("\n---Dữ Liệu Vấn đề 7 ---")
print(df.head(3))


# --- PHẦN 3: XỬ LÝ VẤN ĐỀ 8  ---
# Xác định danh sách cột linh hoạt theo những gì đang có trong bảng
list_time_cols = [col for col in df.columns if col in ['m0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']]
list_id_vars = [col for col in df.columns if col in ['Id', 'Age', 'Weight', 'Firstname', 'Lastname']]

# Tiến hành gộp trục dọc bằng melt an toàn
df = pd.melt(
    df,
    id_vars=list_id_vars,
    value_vars=list_time_cols,
    value_name="PulseRate",
    var_name="sex_and_time"
)

# Sắp xếp hiển thị dữ liệu kết quả chuẩn mực
df = df.sort_values(by=list_id_vars)

print("\n---  Dữ liệu Vấn Đề 8 ---")
print(df.head(10))
```

```python
# =============================================================================
# VẤN ĐỀ 9: Nhiều biến bị lưu gộp trong cột tạm sex_and_time sau khi Melt
# =============================================================================
tmp_df = df["sex_and_time"].str.extract("([a-zA-Z])(\d{2})(\d{2})", expand=True)
tmp_df.columns = ["Sex", "hours_lower", "hours_upper"]

print("\n--- Dữ liệu sau Vấn đề 9 ---")
print(tmp_df.head(5))
```

```python
# =============================================================================
# VẤN ĐỀ 10: Đồng bộ hóa và chuẩn hóa định dạng cột Thời gian (Time)
# =============================================================================
tmp_df["Time"] = tmp_df["hours_lower"] + "-" + tmp_df["hours_upper"]
df = pd.concat([df, tmp_df], axis=1)
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)
df = df.dropna()
df.to_csv('outputcleanup.csv', index=False)

print("\n--- Dữ liệu sau Vấn đề 10 ---")
print(df.head(5))
```

```python
# =============================================================================
# VẤN ĐỀ 11: Khảo sát tỷ lệ khuyết thiếu và điền khuyết Nhịp tim thác đổ
# =============================================================================
df['PulseRate'] = df['PulseRate'].replace('-', np.nan)
df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')

df.sort_values(by=['Id', 'Time'], inplace=True)
df.reset_index(drop=True, inplace=True)

def fill_missing_pulse(row, dataframe):
    if not np.isnan(row['PulseRate']):
        return row['PulseRate']
    idx = row.name
    current_id = row['Id']
    current_sex = row['Sex']
    patient_data = dataframe[dataframe['Id'] == current_id]

    prev_idx = idx - 1 if (idx - 1 in patient_data.index) else None
    next_idx = idx + 1 if (idx + 1 in patient_data.index) else None
    if prev_idx is not None and next_idx is not None:
        p_val = dataframe.loc[prev_idx, 'PulseRate']
        n_val = dataframe.loc[next_idx, 'PulseRate']
        if not np.isnan(p_val) and not np.isnan(n_val):
            return (p_val + n_val) / 2

    prev_idx_2 = idx - 2 if (idx - 2 in patient_data.index) else None
    if prev_idx is not None and prev_idx_2 is not None:
        p_val1 = dataframe.loc[prev_idx, 'PulseRate']
        p_val2 = dataframe.loc[prev_idx_2, 'PulseRate']
        if not np.isnan(p_val1) and not np.isnan(p_val2):
            return (p_val1 + p_val2) / 2

    next_idx_2 = idx + 2 if (idx + 2 in patient_data.index) else None
    if next_idx is not None and next_idx_2 is not None:
        n_val1 = dataframe.loc[next_idx, 'PulseRate']
        n_val2 = dataframe.loc[next_idx_2, 'PulseRate']
        if not np.isnan(n_val1) and not np.isnan(n_val2):
            return (n_val1 + n_val2) / 2

    valid_patient_pulses = patient_data['PulseRate'].dropna()
    if not valid_patient_pulses.empty:
        return valid_patient_pulses.mean()

    gender_mean = dataframe[dataframe['Sex'] == current_sex]['PulseRate'].mean()
    if not np.isnan(gender_mean):
        return gender_mean

    return 70.0

df['PulseRate'] = df.apply(lambda row: fill_missing_pulse(row, df), axis=1)

print("\n--- Dữ liệu sau Vấn đề 11 ---")
print(df.head(5))
```

```python
# =============================================================================
# VẤN ĐỀ 12: Tái thiết lập chỉ mục (Reindex) và xuất file sạch cuối cùng
# =============================================================================
df.dropna(subset=['PulseRate'], inplace=True)
df.reset_index(drop=True, inplace=True)

final_columns = ['Id', 'Firstname', 'Lastname', 'Age', 'Weight', 'Sex', 'Time', 'PulseRate']
df = df[final_columns]

df.to_csv("patient_heart_rate_clean.csv", index=False)

print("\n=======================================================")
print("--- Dự liệu vấn đề 12 ---")
print("=======================================================")
print(df.head(10))
```
