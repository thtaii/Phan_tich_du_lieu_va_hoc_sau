```python
import pandas as pd

df = pd.read_csv('dulieuxettuyendaihoc.csv')

print("CÂU 4: KHẢO SÁT & XỬ LÝ DỮ LIỆU THIẾU CHO CỘT DÂN TỘC (DT)")
print("=====================================================================")

cac_gia_tri_rieng_biet = df['DT'].unique()
print(f"-> Các giá trị riêng biệt có trong cột DT: {cac_gia_tri_rieng_biet}\n")

tan_so_dt = df['DT'].value_counts(dropna=False)
tan_suat_dt = df['DT'].value_counts(dropna=False, normalize=True) * 100

df_khao_sat = pd.DataFrame({
    'Tần số': tan_so_dt,
    'Tần suất (%)': tan_suat_dt
})

print("BẢNG KHẢO SÁT CỘT DÂN TỘC (DT) TRƯỚC KHI XỬ LÝ:")
print(f"{'Giá trị':<12}{'Tần số':<12}{'Tần suất (%)'}")
print("-" * 40)

for gia_tri, row in df_khao_sat.iterrows():

    nhan = str(gia_tri)
    print(f"{nhan:<12}{int(row['Tần số']):<12}{row['Tần suất (%)']:.2f}%")

df['DT'] = df['DT'].fillna(0)

print("\n" + "-" * 40)
print("-> Đã thực hiện thay thế tất cả dữ liệu thiếu bằng giá trị 0.")
so_o_trong_sau_xu_ly = df['DT'].isna().sum()
print(f"-> Số lượng ô trống trong cột DT sau khi xử lý: {so_o_trong_sau_xu_ly}")
```

```python
import pandas as pd


df = pd.read_csv('dulieuxettuyendaihoc.csv')

print("=====================================================================")
print("CÂU 5: KHẢO SÁT & XỬ LÝ DỮ LIỆU THIẾU CHO CỘT TOÁN LỚP 10 (T1)")
print("=====================================================================")

# Lập bảng tần số và tần suất bao gồm cả ô trống NaN
tan_so_t1 = df['T1'].value_counts(dropna=False)
tan_suat_t1 = df['T1'].value_counts(dropna=False, normalize=True) * 100

df_khao_sat_t1 = pd.DataFrame({
    'Tần số': tan_so_t1,
    'Tần suất (%)': tan_suat_t1
})

print("BẢNG KHẢO SÁT CỘT T1 TRƯỚC KHI XỬ LÝ (Hiển thị các dòng tiêu biểu):")
print(f"{'Mức điểm':<12}{'Tần số':<12}{'Tần suất (%)'}")
print("-" * 40)

for diem, row in df_khao_sat_t1.iterrows():
    nhan_diem = str(diem)

    if nhan_diem == 'nan':
        print(f"--> {nhan_diem:<8}{int(row['Tần số']):<12}{row['Tần suất (%)']:.2f}% (Ô TRỐNG)")
    else:
        print(f"{nhan_diem:<12}{int(row['Tần số']):<12}{row['Tần suất (%)']:.2f}%")



# Thực hiện thay thế dữ liệu thiếu bằng phương pháp Mean
mean_t1 = df['T1'].mean()
print("\n" + "-" * 40)
print(f"-> Giá trị trung bình (Mean) tính được của cột T1 là: {mean_t1:.4f}")

df['T1'] = df['T1'].fillna(mean_t1)

print("\n--> Đã thực hiện thay thế tất cả dữ liệu thiếu bằng giá trị Mean.")
so_o_trong_sau_xu_ly = df['T1'].isna().sum()
print(f"--> Số lượng ô trống trong cột T1 sau khi xử lý: {so_o_trong_sau_xu_ly}")
```

```python
#6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
import pandas as pd


df = pd.read_csv('dulieuxettuyendaihoc.csv')
df['DT'] = df['DT'].fillna(0)
df['T1'] = df['T1'].fillna(df['T1'].mean())

print("=====================================================================")
print("CÂU 6: TỰ ĐỘNG XỬ LÝ DỮ LIỆU THIẾU CHO TẤT CẢ CÁC BIẾN ĐIỂM SỐ CÒN LẠI")
print("=====================================================================")

# Định nghĩa danh sách tất cả các cột điểm số học tập và điểm thi đại học
# (Ngoại trừ cột T1 đã được xử lý riêng ở câu 5)
cac_cot_diem_con_lai = [
    'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1', # Các môn lớp 10 còn lại
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2', # Toàn bộ môn lớp 11
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6', # Toàn bộ môn lớp 12
    'DH1', 'DH2', 'DH3'                             # Điểm thi Đại học
]

print(f"{'Tên biến':<10}{'Trạng thái xử lý':<25}{'Giá trị Mean được điền'}")
print("-" * 60)

# Bước 2: Chạy vòng lặp tự động kiểm tra và điền dữ liệu thiếu bằng Mean
for col in cac_cot_diem_con_lai:
    # Kiểm tra xem cột đó có tồn tại thực tế trong file dữ liệu hay không
    if col in df.columns:
        # 1. Đếm xem cột này có bao nhiêu ô bị khuyết thiếu (NaN)
        so_o_trong = df[col].isna().sum()

        # 2. Tính giá trị trung bình (Mean) của riêng cột đó
        mean_cua_cot = df[col].mean()

        # 3. Tiến hành lấp đầy các ô trống bằng giá trị Mean vừa tính
        df[col] = df[col].fillna(mean_cua_cot)

        # In báo cáo tiến trình xử lý ra màn hình
        if so_o_trong > 0:
            print(f"{col:<10}Phát hiện {so_o_trong:<3} ô trống -> Điền Mean | {mean_cua_cot:.4f}")
        else:
            print(f"{col:<10}Dữ liệu đầy đủ (0 ô trống)   | {mean_cua_cot:.4f}")

print("-" * 60)
print("--> [THÀNH CÔNG] Đã làm sạch và xử lý triệt để toàn bộ dữ liệu thiếu cho các biến điểm số!")
```

```python
#7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
print("=====================================================================")
print("CÂU 7: TẠO CÁC BIẾN ĐIỂM TRUNG BÌNH MÔN TBM1, TBM2, TBM3")
print("=====================================================================")

# 1. Tính TBM lớp 10
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10

# 2. Tính TBM lớp 11
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10

# 3. Tính TBM lớp 12
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

# Kiểm tra kết quả
print("Kiểm tra ngẫu nhiên điểm TBM của 5 học sinh đầu tiên:")
print(df[['STT', 'TBM1', 'TBM2', 'TBM3']].head(5))
```

```python
#8.  Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12
print("=====================================================================")
print("CÂU 8: TẠO CÁC BIẾN XẾP LOẠI XL1, XL2, XL3")
print("=====================================================================")

khoang_diem = [0, 5.0, 6.5, 8.0, 9.0, 10.0]
nhan_xep_loai = ['Y', 'TB', 'K', 'G', 'XS']

df['XL1'] = pd.cut(df['TBM1'], bins=khoang_diem, labels=nhan_xep_loai, right=False)
df['XL2'] = pd.cut(df['TBM2'], bins=khoang_diem, labels=nhan_xep_loai, right=False)
df['XL3'] = pd.cut(df['TBM3'], bins=khoang_diem, labels=nhan_xep_loai, right=False)

print("Kiểm tra ngẫu nhiên kết quả TBM và Xếp loại của 5 học sinh đầu tiên:")
print(df[['STT', 'TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head(5))
```

```python
#9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
#10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
#pháp Min-Max Normalization
print("=====================================================================")
print("CÂU 9: CHUYỂN ĐỔI ĐIỂM SANG THANG ĐIỂM 4 MỸ (MIN-MAX NORMALIZATION)")
print("=====================================================================")

# Áp dụng công thức chuẩn hóa Min-Max để quy đổi từ thang 10 sang thang 4
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

# Kiểm tra kết quả hiển thị bằng cách in ra 5 dòng đầu tiên để đối chiếu
print("Bảng đối chiếu điểm hệ 10 và hệ 4 của 5 học sinh đầu tiên:")
print(df[['STT', 'TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head(5))
```

```python
print("=====================================================================")
print("CÂU 10: TẠO BIẾN KẾT QUẢ XÉT TUYỂN (KQXT)")
print("=====================================================================")

# Bước 1: Định nghĩa hàm tính toán kết quả cho từng học sinh
def tinh_kqxt(row):
    khoi = row['KT']
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']

    # Điều kiện cho khối A và A1
    if khoi in ['A', 'A1']:
        diem_xet_tuyen = (dh1 * 2 + dh2 + dh3) / 4
    # Điều kiện cho khối B
    elif khoi == 'B':
        diem_xet_tuyen = (dh1 + dh2 * 2 + dh3) / 4
    # Điều kiện cho các khối khác (C, D,...)
    else:
        diem_xet_tuyen = (dh1 + dh2 + dh3) / 3

    # Trả về 1 nếu Đậu (>= 5.0), trả về 0 nếu Rớt (< 5.0)
    if diem_xet_tuyen >= 5.0:
        return 1
    else:
        return 0

# Bước 2: Áp dụng hàm vừa viết lên từng hàng của DataFrame (axis=1)
df['KQXT'] = df.apply(tinh_kqxt, axis=1)

# Kiểm tra kết quả hiển thị bằng cách xem 10 dòng đầu tiên bao gồm Khối thi và Kết quả
print("Kiểm tra ngẫu nhiên kết quả xét tuyển của 10 học sinh đầu tiên:")
print(df[['STT', 'KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10))

# Thống kê nhanh xem có bao nhiêu bạn đậu, bao nhiêu bạn rớt
print("\nThống kê tổng số lượng Đậu (1) và Rớt (0):")
print(df['KQXT'].value_counts())
```
