```python
from nltk.corpus import stopwords

def check_stopwords_by_lang(lang_list):
    for lang in lang_list:
        try:
            words = stopwords.words(lang)
            print(f"Ngôn ngữ '{lang}': Có {len(words)} stopwords. Ví dụ: {words[:5]}")
        except Exception as e:
            print(f"Không tìm thấy dữ liệu cho ngôn ngữ: {lang}")

# Kiểm tra một số ngôn ngữ phổ biến
check_stopwords_by_lang(['english', 'german', 'spanish', 'vietnamese'])
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Add this line to download the missing resource
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "This is a sample sentence showing how to remove stop words from a text using NLTK."
stop_words = set(stopwords.words('english'))

# Tách từ (Tokenization)
words = word_tokenize(text)

# Loại bỏ stopword
filtered_text = [w for w in words if w.lower() not in stop_words]

print("Văn bản gốc:", words)
print("Văn bản sau khi loại bỏ stopword:", filtered_text)
```

```python
from nltk.corpus import stopwords

# Danh sách stopword mặc định của tiếng Anh
stop_words = set(stopwords.words('english'))

# Giả sử ta muốn giữ lại các từ phủ định như 'not', 'no' phục vụ cho bài toán Sentiment Analysis
words_to_keep = {'not', 'no'}
custom_stop_words = stop_words - words_to_keep

print(f"Số lượng stopword ban đầu: {len(stop_words)}")
print(f"Số lượng stopword sau khi bỏ qua một số từ: {len(custom_stop_words)}")
print("Kiểm tra từ 'not' có trong danh sách mới không:", 'not' in custom_stop_words)
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet

word = "computer"
synsets = wordnet.synsets(word)

if synsets:
    print(f"Từ cần tra cứu: '{word}'")
    # Lấy định nghĩa đầu tiên tìm được
    print("Định nghĩa:", synsets[0].definition())
    # Lấy ví dụ minh họa
    print("Ví dụ đặt câu:", synsets[0].examples())
else:
    print(f"Không tìm thấy định nghĩa cho từ '{word}' trong WordNet.")
```

```python
from nltk.corpus import wordnet

word = "good"
synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for lm in syn.lemmas():
        synonyms.append(lm.name())
        if lm.antonyms():
            antonyms.append(lm.antonyms()[0].name())

# Loại bỏ các phần tử trùng lặp bằng set
print(f"Từ đồng nghĩa của '{word}':", set(synonyms))
print(f"Từ trái nghĩa của '{word}':", set(antonyms))
```

```python
import nltk
nltk.download('tagsets')
nltk.download('tagsets_json') # Add this line to download the missing resource

print("--- 1. Tổng quan và chi tiết của một số tag cụ thể trong Penn Treebank ---")
# Tra cứu tag NN (Danh từ số ít) và VB (Động từ nguyên mẫu)
nltk.help.upenn_tagset('NN')
nltk.help.upenn_tagset('VB')

print("\n--- 2. Sử dụng Regex (Biểu thức chính quy) để tag từ tự động ---")
from nltk import RegexpTagger

patterns = [
    (r'.*ing$', 'VBG'),    # Động từ thêm -ing
    (r'.*ed$', 'VBD'),     # Động từ quá khứ -ed
    (r'.*es$', 'VBZ'),     # Động từ ngôi thứ 3 số ít
    (r'.*s$', 'NNS'),      # Danh từ số nhiều
    (r'^-?[0-9]+(.[0-9]+)?$', 'CD'), # Số đếm
    (r'.*', 'NN')          # Mặc định là Danh từ nếu không khớp mẫu nào
]

regexp_tagger = RegexpTagger(patterns)
sample_text = ["playing", "walked", "books", "12.5", "table"]
print("Kết quả gán tag bằng Regex:")
print(regexp_tagger.tag(sample_text))
```

```python
from nltk.corpus import wordnet

# Định nghĩa 2 danh từ dưới dạng synset
noun1 = wordnet.synset('dog.n.01')
noun2 = wordnet.synset('cat.n.01')

# Đo lường mức độ tương đồng dựa trên đường đi ngắn nhất (Path Similarity)
similarity = noun1.path_similarity(noun2)

print(f"Độ tương đồng giữa '{noun1.name()}' và '{noun2.name()}' là: {similarity}")
```

```python
from nltk.corpus import wordnet

verb1 = wordnet.synset('run.v.01')
verb2 = wordnet.synset('walk.v.01')

similarity = verb1.path_similarity(verb2)

print(f"Độ tương đồng giữa '{verb1.name()}' và '{verb2.name()}' là: {similarity}")
```

```python
import nltk
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder

# Tải và lấy từ trong tác phẩm Macbeth
words = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')

# Tạo bộ tìm kiếm cụm 2 từ (Bigrams)
finder = BigramCollocationFinder.from_words(words)

# Lọc bỏ các dấu câu hay từ quá ngắn nếu muốn, hoặc in ra các cụm xuất hiện nhiều nhất
bigram_measures = BigramAssocMeasures()
top_15_bigrams = finder.nbest(bigram_measures.raw_freq, 15)

print("15 kết hợp từ (Bigrams) xuất hiện phổ biến nhất trong Macbeth:")
for bigram in top_15_bigrams:
    print(' '.join(bigram))
```
